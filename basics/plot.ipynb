# Plotting

**Draw a plot**
* lty = linetype, possible values: 
    * 0: blank
    * 1: solid
    * 2: dashed
    * 3: dotted
    * 4: dotdash
    * 5: longdash
    * 6: twodash
* type = type of graph
    * “p”: Points.
    * “l”: Lines.
    * “b”: Both.
    * “c”: The lines part alone of “b”
    * “o”: Both “overplotted”
    * “h”: Histogram like (or high-density) vertical lines.
    * “n”: No plotting.

```r
z <- c(4, 2, 8, 9, 7, 5, 2, 1)
plot(z) # everything default
```

```r
plot(z,
     type = "l",
     col = "blue",
     lty = 2,
     main = "Haupttitel",
     xlab = "Ein paar Zahlen",
     ylab = "Andere Zahlen"
)
```

Add lines to plot.
v = vertical line
h = horizontal line
a,b = slope

```r
plot(z)
abline(v=3, col="green", lty=1)
abline(h=4, col="red", lty=3)
abline(a = 1, b = 2, col="brown", lty=5) # slope: y=2x+1   
```
