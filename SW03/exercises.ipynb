### Aufgabe 3.3 <a name="aufgabe-3-3"></a>

```r
income <- read.table(file="./data/income.dat", header=TRUE)
head(income)
```

```r
iq <- income[,1]
anzahl_jahre_schule <- income[,2]
einkommen <- income[,3]
plot(iq,
einkommen,
type = "p",
xlab = "IQ",
ylab = "Einkommen",
col = "lightblue"
)
abline(lm(einkommen ~ iq),
col = "orange",
lwd = 3)
```

```r
plot(anzahl_jahre_schule,
    einkommen,
    type="p",
     xlab = "Anzahl Jahre Schulbildung",
    ylab="Einkommen",
    col = "lightblue"
)
abline(lm(einkommen ~ anzahl_jahre_schule),
col = "orange",
lwd = 3)
```

```r
lm(einkommen ~ anzahl_jahre_schule)
```
