# Streudiagramm 
**(Scatter plot)**

```r
wein <- c(2.8, 3.2, 3.2, 3.4, 4.3, 4.9, 5.1, 5.2, 5.9, 5.9, 6.6, 8.3, 12.6, 15.1, 25.1, 33.1, 75.9, 75.9)
mort <- c(6.2, 9.0, 7.1, 6.8, 10.2, 7.8, 9.3, 5.9, 8.9, 5.5, 7.1, 9.1, 5.1, 4.7, 4.7, 3.1, 3.2, 2.1)

plot(wein, mort,
xlab = "Weinkonsum (Liter pro Jahr)",
ylab = "MortalitÃ¤t",
col = "blue",
pch = 20
)
```

## Regressionsgerade

Regressionsgerade:
y = a + bx

ausrechnen mit "lm" 

```r
lm(wein ~ mort) # 63.476 ist a, -7.306 ist b
```

### Streudiagramm mit Regressionsgerade

```r
seiten <- seq(50, 500, 50)
preis <- c(6.4, 9.5, 15.6, 15.1, 17.8, 23.4, 23.4, 22.5,
26.1, 29.1)
plot(seiten, preis,
col = "orange",
pch = 19,
xlab = "Seitenzahl",
ylab = "Buchpreis"
)
abline(lm(preis ~ seiten), col = "deepskyblue")
```
